```python
import pandas as pd
import json

# Read the JSON file
with open('/Users/andysheung/Appcara/DL Data/dl_arr.json', 'r') as file:
    dl_data = json.load(file)

# Assuming df is your Pandas DataFrame

# Calculate the average temperature for the last 1 hour
#df['Average_Temp'] = df.filter(like='_Temp').mean(axis=1)
print(df)
# Define the racks and aisles for the table
racks = {
    "Heater rack": ["Heater rack - Cold aisle", "Heater rack - Hot aisle(left)", "Heater rack - Hot aisle(right)"],
    "Normal rack": ["Normal rack - Cold aisle", "Normal rack - Hot aisle(left)", "Normal rack - Hot aisle(right)"]
}

aisles = ["top", "mid", "low"]

# Create a dictionary to store the average temperatures for each rack and aisle
average_temps = []

for rack_type, rack_names in racks.items():
    for rack_name in rack_names:
        for i, aisle in enumerate(aisles):
            key = f"{rack_type} - {aisle}"
            dl_numbers = dl_data.get(rack_name)[i]
            temps = df.loc[df.index.isin(dl_numbers), 'Average_Temp']
            average_temp = temps.mean() if not temps.empty else 0
            average_temps.append((key, round(average_temp, 2)))

# Print or use the list of average temperatures
#print(average_temps)
```
